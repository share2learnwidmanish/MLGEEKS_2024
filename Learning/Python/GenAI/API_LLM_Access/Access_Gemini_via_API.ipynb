```python
import google.generativeai as genai
#import genai

genai.configure(api_key="API_KEY")  # Specify API version

model_name = "gemini-1.5-pro-creative"  # Replace with the actual model name (if available)

# Define the prompt and safety settings
prompt = "Write a poem about nature."
safety_settings = genai.safety_types.SafetySettingOptions(
    profanity_filter=True,  # Enable profanity filtering
    hate_speech_filter=True   # Enable hate speech filtering
)

# Generate content
response = model.generate_content(
    contents=[prompt], safety_settings=safety_settings
)

# Handle the generation response
if response.error is None:
    generated_text = response.content.text
    print(f"Generated poem:\n{generated_text}")
else:
    print(f"Error during generation: {response.error}")
```
